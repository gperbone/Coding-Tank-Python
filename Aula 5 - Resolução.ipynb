Faça uma função que recebe dois parâmetros: linhas e colunas e desenhe uma tabela usando os caracteres (-, |, + e espaços) com essas dimensões, exemplo:

linhas: 3
colunas: 5

    -----------
    | | | | | |
    |-+-+-+-+-|
    | | | | | |
    |-+-+-+-+-|
    | | | | | |
    -----------

```python
def imprime_tabela(linhas, colunas):
    print('-' * (colunas*2 + 1))
    i = 1
    while i < linhas:
        print("| " * (colunas + 1))
        print("|-" + "+-" * (colunas - 1) + "|")
        i += 1
    
    print("| " * (colunas + 1))
    print('-' * (colunas*2 + 1))

imprime_tabela(3,5)
```

Faça uma função que recebe um texto e um letra e retorne a quantidade daquela letra naquele texto (ignore diferenças de capitalização, 'A' e 'a' são a mesma letra)

```python
def cont_letra(texto, letra):
    i = len(texto) - 1
    cont = 0

    while i >= 0:
        if texto[i].lower() == letra.lower():
            cont += 1
        i -= 1
    
    return cont

print(cont_letra('Ola mundo, estou aqui!', 'a'))
print(cont_letra('Ola mundo, estou aqui!', 'o'))
```

Faça uma função que recebe um inteiro n e retorna uma lista com os n primeiros elementros da sequência de fibonacci

```python
def fibo(n):
    if n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fibo(n-1) + fibo(n-2)

def sequencia_fibo(n):
    i = 1
    lista = []

    while i <= n:
        lista.append(fibo(i))
        i += 1

    return lista

sequencia_fibo(12)
```

Faça uma função maxLista que faz a mesma coisa que a funcão _max_ faz ao receber uma lista

```python
def max_lista(lista):
    max = lista[0]
    i = 1

    while i < len(lista):
        if lista[i] > max:
            max = lista[i]

        i += 1

    return max

max([1,3,4,6,534,7,545,90,987,234,465,675])
```

Faça uma função que calcule fatoriais utilizando recursão:

```python
def fatorial(n):
    if n == 0:
        return 1
    else:
        return n * fatorial(n-1)

fatorial(3)
```
