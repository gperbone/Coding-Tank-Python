Faça uma função que recebe dois parâmetros: linhas e colunas e desenhe uma tabela usando os caracteres (-, |, + e espaços) com essas dimensões, exemplo:

linhas: 3
colunas: 5

    -----------
    | | | | | |
    |-+-+-+-+-|
    | | | | | |
    |-+-+-+-+-|
    | | | | | |
    -----------

```python
def cria_tabela(n_linhas, n_colunas):
    
    if type(n_linhas) != int or type(n_colunas) != int:
        return 'Número de linhas e colunas deve ser positivo'
    elif n_linhas <= 0 or n_colunas <= 0:
        return 'Número de linhas e colunas deve ser positivo'
    
    print('-' + 2*n_colunas*'-')
    
    i = 1
    
    while i < n_linhas:
        
        print('|' + n_colunas*' |')
        
        print('|'+ (n_colunas - 1)*'-+' + '-|')
        
        i += 1
        
    print('|' + n_colunas*' |')
    print('-' + 2*n_colunas*'-')

    
```

```python
cria_tabela(3, 10)
```

Faça uma função que recebe um texto e um letra e retorne a quantidade daquela letra naquele texto (ignore diferenças de capitalização, 'A' e 'a' são a mesma letra)

```python
def conta_caracter(texto, caracter):
    
    texto_aux = texto.lower()
    
    contador = 0
    
    i = 0
    
    while i < len(texto_aux):
        if texto_aux[i] == caracter:
            contador += 1
            
        i+= 1
            
    return contador
        
```

```python
conta_caracter('O Brasil é um país muito bonito', 'o')
```

Faça uma função que recebe um inteiro n e retorna uma lista com os n primeiros elementros da sequência de fibonacci

```python
# Funcao que calcula o n-esimo termo de fibonacci
def fibonacci(n):
    
    if n == 1:
      return 0
    elif n == 2:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
    
def lista_fibonacci(n):
    
    lista_fibo = []
    
    i = 1
    
    while i <= n:
        lista_fibo.append(fibonacci(i))

        i += 1
        
    return lista_fibo
    
```

```python
lista_fibonacci(10)
```

Faça uma função maxLista que faz a mesma coisa que a funcão _max_ faz ao receber uma lista

```python
def max_lista(lista):
    
    if type(lista[0]) not in (int, float):
        return 'Lista deve ser numérica'
    
    maximo = lista[0]
    
    i = 1
    
    while i < len(lista):
        if type(lista[i]) not in (int, float):
            return 'Lista deve ser numérica'
        elif lista[i] > maximo:
            maximo = lista[i]
            
        i += 1
        
    return maximo
```

```python
max_lista([101, 2, 3, 5, 100])
```

```python
max_lista(['a', 101, 2, 3, 5, 100])
```

```python
max_lista([101, 'b', 2, 3, 5, 100])
```

Faça uma função que calcule fatoriais utilizando recursão:

```python
def calcula_fatorial(n):
    
    if type(n) != int:
        return 'Insira um número inteiro'
    elif n == 1:
        return 1
    else:
        return n * calcula_fatorial(n - 1)
    
```

```python
calcula_fatorial('a')
```

```python
calcula_fatorial(10.5)
```

```python
i = 1
while i < 10:
    print(f'{i}! = {calcula_fatorial(i)}')
    i += 1
```
